```python
import pandas as pd
data=pd.read_csv("data/clean_data.csv")
```

```python
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig,axs=plt.subplots(2,2,figsize=(10,8))
sns.histplot(data=data,x="average",kde=True,color="b",ax=axs[0][0])
sns.histplot(data=data,x="average",hue="gender",ax=axs[0][1])

sns.histplot(data=data,x="total_score",kde=True,color="b",ax=axs[1][0])
sns.histplot(data=data,x="total_score",hue="gender",ax=axs[1][1])
```

```python
fig,axs=plt.subplots(1,3,figsize=(10,3))

sns.histplot(data=data,x="average",kde=True,hue="lunch",ax=axs[0])
sns.histplot(data=data[data.gender=="female"],kde=True,x="average",hue="lunch",ax=axs[1])
sns.histplot(data=data[data.gender=="male"],kde=True,x="average",hue="lunch",ax=axs[2])
```

```python
fig,axs=plt.subplots(1,3,figsize=(10,3))

sns.histplot(data=data,x="average",kde=True,hue="parental_level_of_education",ax=axs[0])
sns.histplot(data=data[data.gender=="female"],kde=True,x="average",hue="parental_level_of_education",ax=axs[1])
sns.histplot(data=data[data.gender=="male"],kde=True,x="average",hue="parental_level_of_education",ax=axs[2])
```

```python
fig,axs=plt.subplots(1,3,figsize=(10,3))

sns.histplot(data=data,x="average",kde=True,hue="race_ethnicity",ax=axs[0])
sns.histplot(data=data[data.gender=="female"],kde=True,x="average",hue="race_ethnicity",ax=axs[1])
sns.histplot(data=data[data.gender=="male"],kde=True,x="average",hue="race_ethnicity",ax=axs[2])
```
